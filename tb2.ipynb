```julia
function calc_Htb(Nx)
    mat_Htb=zeros(Float64,Nx,Nx)
    t=1.0
    for i=1:Nx
        for dx=-1:1
            j=i+dx 
         j+=ifelse(j>Nx,-Nx,0)
         j+=ifelse(j<1,Nx,0)
                
            if abs(dx)==1
                mat_Htb[i,j]=-t
            elseif dx==0
                mat_Htb[i,j]=0.0
            end
        end
    end
    
    return mat_Htb
    
end 
```

```julia
using Plots
gr()
using LinearAlgebra
Nx=100
mat_Htb=calc_Htb(Nx)
ε,c=eigen(mat_Htb)
plot(ε[:])
```

```julia
function calc_Htb(Nx,μ)
    mat_Htb=zeros(Float64,Nx,Nx)
    t=1.0
    for i=1:Nx
        for dx=-1:1
            j=i+dx 
         j+=ifelse(j>Nx,-Nx,0)
         j+=ifelse(j<1,Nx,0)
                
            if abs(dx)==1
                mat_Htb[i,j]=-t
            elseif dx==0
                mat_Htb[i,j]=-μ
            end
        end
    end
    
    return mat_Htb #functionの出力値
    
end 
```

```julia
using Plots
gr()
using LinearAlgebra
Nx=100
μ=-1.5
mat_Htb=calc_Htb(Nx,μ)
ε,c=eigen(mat_Htb)
plot(ε[:])
```

```julia
#forループはfunction内で書いたほうが速くなるので書き直した。
function count(ε)
    
    N=0
    for i in ε #julianote.pdf forループサンプル２参照
        if i>0
            break #exit
        end
        N+=1
    end
    
    return N
    
end
```
